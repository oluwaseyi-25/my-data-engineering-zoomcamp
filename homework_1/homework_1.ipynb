```python
zones = pd.read_csv("taxi_zone_lookup.csv")
zones.to_sql(con=engine, name='zones', if_exists='replace')
```

```python
green_trips = pd.read_parquet('green_tripdata_2025-11.parquet')
green_trips.to_sql(name='green_trip_data', con=engine, if_exists='replace')
```
